```python
import seaborn as sns

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.set_style('darkgrid')
sns.barplot(x='species', y='body_mass_g', data=penguins);
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            ci=68
           );#CI-confidence interval default is 95#When ci is set to 68, it indicates a 68% confidence interval. 
#This means that the error bars on the plot will represent the 68% confidence interval around the mean body mass of penguins for each species.
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            ci=None
           );
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            estimator=np.std,ci=None
           );#Displays only the std of each type
```

```python
penguins.groupby('species').body_mass_g.describe()
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            estimator=np.max
           );#estimator=np.max specifies that for each category on the x-axis (each species of penguin),
            #the maximum value of body mass will be used to determine the height of the bar.
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            hue='sex'
           );
```

```python
sns.barplot(x='species', y= 'body_mass_g', data=penguins,order=['Gentoo', 'Chinstrap', 'Adelie']);#Changing the order
```

```python
sns.barplot(x='species', y='body_mass_g', data=penguins, color='c');
```

```python
sns.barplot(x='species', y='body_mass_g', data=penguins, color='c',errcolor='darkslateblue',errwidth=8);
```

```python
sns.barplot(x='species', y='body_mass_g', data=penguins,palette='hot');
```

```python
sns.barplot(x='species', y='body_mass_g', data=penguins, ci=None,color='aliceblue',edgecolor='mediumblue',lw=3);
```

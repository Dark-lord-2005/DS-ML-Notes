```python
Distribution plots/distplot
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
sns.distplot(cars.weight);
```

```python
#Distplot combines histogram,kde plot & rugplot(turned of by defualt) in a single plot
#To print the histogram alone:
sns.distplot(cars.weight,kde=False);
```

```python
#We can also turn of histo and print the kde alone
sns.distplot(cars.weight,hist=False);
```

```python
sns.distplot(cars.weight,hist=False,kde=False);#No graph will be printed except x and y axis
```

```python
sns.distplot(cars.weight,rug=True);
```

```python
#The default no of bins is 10 but can be changed as follows:(Bins are equally placed from min to max val of data
sns.distplot(cars.weight, bins=5);  
```

```python
#Bins of different sizes can be used by passing appropriate lists with values
sns.distplot(cars.weight, bins=[1500, 2000, 2500, 3500, 5200]);
```

```python
#FIT - fits a distrubution to our graph fromn the scimpy library:
from scipy.stats import skewnorm

sns.distplot(cars.weight, fit=skewnorm,kde=False);
```

```python
from scipy.stats import norm#What we import is important

sns.distplot(cars.weight, fit=norm, kde=False);
```

```python
#Kwargs add design to the output:
sns.distplot(cars.weight, kde=False, hist_kws={'alpha': 1});#Improving tranparency of bars of hist
```

```python
sns.distplot(cars.weight, kde=False, 
             hist_kws={'alpha': 1, 'histtype': 'stepfilled'});#Outline alone if histtype is step
```

```python
sns.distplot(cars.weight, kde_kws={'lw': 5});#To make the line of kde plot more thicker
```

```python
sns.kdeplot(cars.weight, lw=5);  
```

```python
#Working with the bandwidth
sns.distplot(cars.weight, kde_kws={'bw': 1});
```

```python
#Working with the rug plot
#Changing the height of the rug
sns.distplot(cars.weight, rug=True, rug_kws={'height': 0.15});
```

```python
#Changing color of the entire plot:
sns.distplot(cars.weight, rug=True, rug_kws={'height': 0.15},color='g');
```

```python
#To give different colours
sns.distplot(cars.weight, rug=True, rug_kws={'height': 0.15,'color':'r'},kde_kws={'color':'k'});
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
KDE Plots
```

```python
cars.head()
```

```python
sns.kdeplot(cars.weight);
```

```python
sns.kdeplot(cars.acceleration);
```

```python
sns.kdeplot(cars.acceleration,shade=True);
```

```python
sns.kdeplot(cars.acceleration,bw=1);
```

```python
sns.kdeplot(cars.acceleration,bw=50);
```

```python
#if instead of pdf we want cdf then
```

```python
sns.kdeplot(cars.acceleration,cumulative=True);
```

```python
#Working with 2D datas:Uses joint pdf 
```

```python
sns.kdeplot(x=cars.horsepower,y=cars.mpg)
```

```python
#having more number of rings
sns.kdeplot(x=cars.horsepower,y=cars.mpg,n_levels=20)
```

```python
#having the regions shaded
sns.kdeplot(x=cars.horsepower,y=cars.mpg,shade=True)
```

```python
#To see what those different densities respresent:
sns.kdeplot(x=cars.horsepower,y=cars.mpg,shade=True,cbar=True)
```

```python
sns.set_style('darkgrid')
```

```python
sns.scatterplot(x=cars.horsepower,y=cars.mpg, color='black', s=20)
plt.xlim(0, 260)
plt.ylim(0, 55)
plt.xlabel('Horsepower', fontsize=16)
plt.ylabel('MPG', fontsize=16)
plt.tight_layout();
```

```python
sns.kdeplot(x=cars.horsepower,y=cars.mpg)
sns.scatterplot(x=cars.horsepower,y=cars.mpg, color='black', s=20,zorder=10)
plt.xlim(0, 260)
plt.ylim(0, 55)
plt.xlabel('Horsepower', fontsize=16)
plt.ylabel('MPG', fontsize=16)
plt.tight_layout();
```

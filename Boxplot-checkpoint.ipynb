```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
#import pandas as pd
#pd.read_csv('C:\Users\veera\AppData\Local\seaborn\seaborn\Cache\mpg')(To load it from system)
```

```python
sns.set_style('whitegrid')
cars.cylinders.value_counts()
```

```python
#Consider only even number of cylinder counts
cars = cars[cars.cylinders.isin([4,6,8])]
sns.boxplot(cars.mpg);
```

```python
cars.mpg.describe()
```

```python
#As it can be seen,25%=17,75%=29,diff is 23% and the extension is till 9 and 46 as they both lie inside 1.5*..there is no outlier
```

```python
sns.boxplot(x=cars.origin, y=cars.mpg);#sns.boxplot(x='origin', y='mpg', data=cars);-can also be used
```

```python
#The diamond values are the outlier ones
```

```python
sns.boxplot(x='origin', y='mpg', hue='cylinders', data=cars);
#Using hue
```

```python
#BY MODEL YEAR
cars.model_year.describe()
```

```python
sns.boxplot(cars.model_year)
```

```python
cars['newer_model'] = cars.model_year > 76
sns.boxplot(x='origin', y='mpg', hue='newer_model', data=cars);
```

```python
#Changing the order of occurence of origin:
sns.boxplot(y=cars.mpg,x=cars.origin,order=['japan','usa','europe'])
```

```python
#To change the order of hue:
sns.boxplot(x='mpg', y='origin', hue='newer_model',
            data=cars,
            order=['japan', 'europe', 'usa'],
            hue_order=[True, False]
           );
```

```python
#Changing the color of the boxes:
sns.boxplot(x='mpg', y='origin', data=cars, color='r');
```

```python
sns.boxplot(x='mpg', y='origin', hue='newer_model',
            data=cars,
            color='r'
           );#False value will get the lighter version of the given color
```

```python
sns.boxplot(x='mpg', y='origin', data=cars, width=0.5,linewidth=5); 
```

```python
#The property whis is the that deteremines wether a point is anoutlier..eg standard whis is 1.5
sns.boxplot(x='mpg', y='origin', data=cars,whis=2);
```

```python
#fliersize is used to change the size of the outliers..Default value is 5
sns.boxplot(y='origin',x='mpg',data=cars,fliersize=2)
plt.savefig("seaborn_plot.png")
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
#Filtering the data set and taking only 100 values for a better understanding
diamonds = diamonds[diamonds.cut.isin(['Premium', 'Good']) & diamonds.color.isin(['D', 'F', 'J'])].sample(n=100, random_state=22)
diamonds.shape
```

```python
sns.set_style('dark')
sns.scatterplot(x=diamonds.carat,y=diamonds.price);
```

```python
sns.scatterplot(x='carat',y='price',data=diamonds,color='g')
```

```python
#Seperating thepoints based on hue
sns.scatterplot(x='carat', y='price', hue='cut',data=diamonds,hue_order=['Good', 'Premium']);
```

```python
#Changing color 
sns.scatterplot(x='carat', y='price', hue='cut',
                data=diamonds,
                palette=['purple', '#55CCCC'],hue_order=['Good', 'Premium']
               );
```

```python
#Changing the order
sns.scatterplot(x='carat', y='price', hue='cut',data=diamonds,palette = ['purple', '#55CCCC'],hue_order=['Good', 'Premium']);
```

```python
sns.scatterplot(x='carat', y='price', hue='depth',
                data=diamonds
               );#For continuous variables
```

```python
#Changing the style of points:
sns.scatterplot(x='carat', y='price', style='cut',data=diamonds,s=100);
```

```python
#Changing size ofpoints:
sns.scatterplot(x='carat', y='price', size='cut',data=diamonds, sizes=[150, 50],size_order=['Good', 'Premium']);
```

```python
sns.scatterplot(x='carat', y='price', hue='cut', style='cut',data=diamonds,s=100);
```

```python
sns.scatterplot(x='carat', y='price', 
                hue='cut', 
                style='color',
                data=diamonds,
                s=100
               );
```

```python
sns.scatterplot(x='carat', y='price', 
                hue='cut', 
                style='color',
                size='depth',
                data=diamonds);
```

```python
#Changing transparency
sns.scatterplot(x='carat', y='price', data=diamonds,s=100, alpha=0.6);
```

```python
#Changing the markers
sns.scatterplot(x='carat', y='price', 
                data=diamonds, 
                s=200, 
                marker='*',
                edgecolor='black'
               );
```

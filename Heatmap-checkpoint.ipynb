```python
import seaborn as sns

from matplotlib import pyplot as plt
import numpy as np
cars = sns.load_dataset('mpg')
cars.head()
```

```python
cars.groupby('origin').cylinders.value_counts()
```

```python
(cars.groupby('origin').cylinders.value_counts().unstack().fillna(0))
```

```python
origin_cyl = (cars.groupby('origin').cylinders.value_counts().unstack().fillna(0))
sns.heatmap(origin_cyl);
```

```python
test_array = np.array([[1, 6], [3, 4]])
sns.heatmap(test_array);
```

```python
sns.heatmap(origin_cyl.T);#Transposes the heatmap
```

```python
#To change colors:
sns.heatmap(origin_cyl, cmap="Blues");
```

```python
#To have a mix of colors
sns.heatmap(origin_cyl, cmap='RdBu');
```

```python
sns.heatmap(origin_cyl, cmap='RdBu', center=60);#WE can modify centerwhich holds the lightestcolor
```

```python
sns.heatmap(origin_cyl,cmap='Greens',vmax=50,vmin=0)#(Value max and Valuemin)
```

```python
#Annotations-prints the value along with the colors
sns.heatmap(test_array, cmap='Blues',annot=True)
```

```python
sns.heatmap(origin_cyl, cmap='Blues', annot=True, fmt=".0f", annot_kws={'fontsize': 16,'fontweight': 'bold','fontfamily': 'serif','color': 'black' });
```

```python
#Leaving gaps b/w squares
sns.heatmap(origin_cyl, cmap='RdBu', linewidth=1);
```

```python
sns.heatmap(origin_cyl, cmap='RdBu', linewidth=1,linecolor='black');
```

```python
labels = {'y':['Europe','USA','Japan'],'x':['3','4','5','6','8']}
sns.heatmap(origin_cyl, cmap='RdBu',xticklabels=labels['x'],yticklabels=labels['y']);#Default
```

```python
labels = {'y':['Europe','USA','Japan'],'x':['3','4','5','6','8']}
sns.heatmap(origin_cyl, cmap='RdBu',xticklabels=labels['x'],yticklabels=False);#Default
```

```python
#Different shape
sns.heatmap(origin_cyl, cmap='RdBu',square=True);
```

```python
#Countplot is similar tobar plot
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
diamonds = diamonds[diamonds.clarity.isin(['SI1', 'VS2'])]
diamonds.head()
```

```python
sns.set_style('darkgrid')
sns.countplot(x='color', data=diamonds);
```

```python
color_order=['J', 'I', 'H', 'G', 'F', 'E', 'D']
sns.countplot(x='color', data=diamonds, order=color_order);
```

```python
sns.countplot(x='color', data=diamonds,
              order=diamonds.color.value_counts().index[::-1]  #remove [::-1] for descending
             ); #Printing in ascending or descending order
```

```python
sns.countplot(x='color', data=diamonds, hue='clarity', hue_order=['SI1', 'VS2']);#Using hue
```

```python
sns.countplot(x='color', data=diamonds, palette='twilight');#Using different colour scheme
```

```python
sns.countplot(x='color', data=diamonds,palette='twilight',
              lw=4, ec='black', hatch='|'
             );
```
